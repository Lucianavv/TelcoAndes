```python
# Ordenar los datos por usuario y mes para calcular valores del mes anterior
fact_table.sort_values(by=["user_id", "month_year"], inplace=True)

# Crear nuevas características basadas en el uso y los excedentes del mes actual
fact_table["extra_minutes"] = (fact_table["duration"] - fact_table["minutes_included"]).clip(lower=0)
fact_table["extra_mb"] = (fact_table["mb_used"] - fact_table["mb_per_month_included"]).clip(lower=0)
fact_table["extra_messages"] = (fact_table["message_count"] - fact_table["messages_included"]).clip(lower=0)

# Calcular el costo extra del mes actual
fact_table["extra_cost"] = (fact_table["extra_minutes"] * fact_table["usd_per_minute"]) + \
                           (fact_table["extra_mb"] * fact_table["usd_per_gb"] / 1024) + \
                           (fact_table["extra_messages"] * fact_table["usd_per_message"])

# Calcular el costo total del mes actual
fact_table["total_cost"] = fact_table["usd_monthly_pay"] + fact_table["extra_cost"]

# Crear variables del mes anterior para predecir el siguiente mes
fact_table["prev_total_cost"] = fact_table.groupby("user_id")["total_cost"].shift(1)
fact_table["prev_extra_minutes"] = fact_table.groupby("user_id")["extra_minutes"].shift(1)
fact_table["prev_extra_mb"] = fact_table.groupby("user_id")["extra_mb"].shift(1)
fact_table["prev_extra_messages"] = fact_table.groupby("user_id")["extra_messages"].shift(1)

# Eliminar registros sin historial previo (primer mes de cada usuario)
fact_table.dropna(subset=["prev_total_cost"], inplace=True)
```
