# Visualizaciones

```python
# Convertir fechas a datetime
calls["call_date"] = pd.to_datetime(calls["call_date"])
internet["session_date"] = pd.to_datetime(internet["session_date"])
messages["message_date"] = pd.to_datetime(messages["message_date"])
users["reg_date"] = pd.to_datetime(users["reg_date"])
users["churn_date"] = pd.to_datetime(users["churn_date"])

# 1. Distribuci√≥n de duraci√≥n de llamadas
plt.figure(figsize=(8, 5))
sns.histplot(calls["duration"], bins=30, kde=True)
plt.title("Distribuci√≥n de Duraci√≥n de Llamadas")
plt.xlabel("Duraci√≥n (min)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=calls["duration"])
plt.title("Boxplot de Duraci√≥n de Llamadas")
plt.xlabel("Duraci√≥n (min)")
plt.show()
```

```python
# 2. Distribuci√≥n de consumo de datos
plt.figure(figsize=(8, 5))
sns.histplot(internet["mb_used"], bins=30, kde=True)
plt.title("Distribuci√≥n de Consumo de Datos")
plt.xlabel("MB Usados")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=internet["mb_used"])
plt.title("Boxplot de Consumo de Datos")
plt.xlabel("MB Usados")
plt.show()
```

MB usados en parecido a la duracion en minutos

```python
# 3. Cantidad de mensajes enviados por usuario
messages_per_user = messages["user_id"].value_counts()
plt.figure(figsize=(8, 5))
sns.histplot(messages_per_user, bins=30, kde=True)
plt.title("Distribuci√≥n de Mensajes Enviados por Usuario")
plt.xlabel("Cantidad de Mensajes")
plt.ylabel("Frecuencia")
plt.show()
```

Se notan posibles outliers mayores a 1000, hay algunos usuarios que textean muchisimo

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=messages_per_user)
plt.title("Boxplot de Mensajes Enviados por Usuario")
plt.xlabel("Cantidad de Mensajes")
plt.show()
```

Valores extremos de cantidad de mensajes mayores a 1000

```python
plt.figure(figsize=(8, 5))
sns.histplot(users["age"], bins=20, kde=True)
plt.title("Distribuci√≥n de Edad de los Usuarios")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# 5. Usuarios por ciudad
plt.figure(figsize=(10, 5))
users["city"].value_counts().nlargest(10).plot(kind='bar')
plt.title("Top 10 Ciudades con M√°s Usuarios")
plt.xlabel("Ciudad")
plt.ylabel("Cantidad de Usuarios")
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Comparaci√≥n de planes
plt.figure(figsize=(6, 4))
sns.barplot(x=plans["plan_name"], y=plans["usd_monthly_pay"])
plt.title("Costo Mensual por Plan")
plt.xlabel("Plan")
plt.ylabel("Costo Mensual ($)")
plt.show()
```
